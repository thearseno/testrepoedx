```python
%sql DROP TABLE IF EXISTS SPACEXTABLE;
```

```python
%sql create table SPACEXTABLE as select * from SPACEXTBL where "Date" is not null
```

## Tasks

### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
df_check = pd.read_sql_query(
    "SELECT distinct Launch_Site FROM SPACEXTABLE;",
    con
)

df_check
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'KSC' 


```python
df_check = pd.read_sql_query(
    """SELECT * FROM SPACEXTABLE
    WHERE LAUNCH_SITE LIKE 'KSC%'
    LIMIT 5;""",
    con
)

df_check
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
df_check = pd.read_sql_query(
    """SELECT Customer, sum(PAYLOAD_MASS__KG_) as total_mass 
    FROM SPACEXTABLE
    WHERE Booster_Version = 'NASA (CRS)'
    GROUP BY 1
    ORDER BY 2 DESC;""",
    con
)

df_check
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
df_check2 = pd.read_sql_query(
    """SELECT Booster_Version, avg(PAYLOAD_MASS__KG_) as avg_mass 
    FROM SPACEXTABLE
    WHERE Booster_Version = 'F9 v1.1'
    GROUP BY 1
    ORDER BY 2 DESC;""",
    con
)

df_check2
```

### Task 5

##### List the date where the succesful landing outcome in drone ship was acheived.


_Hint:Use min function_ 


```python
df_check3 = pd.read_sql_query(
    """SELECT Mission_Outcome, min(Date) as frst_date
    FROM SPACEXTABLE
    WHERE Booster_Version = 'F9 v1.1'
    GROUP BY 1
    ORDER BY 2 DESC;""",
    con
)

df_check3
```

### Task 6

##### List the names of the boosters which have success in ground pad  and have payload mass greater than 4000 but less than 6000


```python
df_check4 = pd.read_sql_query(
    """SELECT Booster_Version
    FROM SPACEXTABLE
    WHERE Landing_Outcome = 'Success (ground pad)'
    AND PAYLOAD_MASS__KG_ > 4000 
    AND PAYLOAD_MASS__KG_ < 6000""",
    con
)

df_check4
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
df_check5 = pd.read_sql_query(
    """
    SELECT
        Mission_Outcome,
        COUNT(DISTINCT (Date || "Time (UTC)")) AS Total_Missions
    FROM SPACEXTABLE
    GROUP BY Mission_Outcome
    """,
    con
)

df_check5

```

### Task 8



##### List all the booster_versions that have carried the maximum payload mass. Use a subquery.


```python
df_check6 = pd.read_sql_query(
    """
SELECT DISTINCT Booster_Version
FROM SPACEXTABLE
WHERE PAYLOAD_MASS__KG_ = (
    SELECT MAX(PAYLOAD_MASS__KG_)
    FROM SPACEXTABLE
);

    """,
    con
)

df_check6
```

### Task 9


##### List the records which will display the month names, succesful landing_outcomes in ground pad ,booster versions, launch_site for the months in year 2017
**Note: SQLLite does not support monthnames. So you need to use substr(Date,6,2) for month, substr(Date,9,2) for date, substr(Date,0,5),='2017' for year.**


```python
df_check7 = pd.read_sql_query(
    """
    SELECT
        CASE substr(Date, 6, 2)
            WHEN '01' THEN 'January'
            WHEN '02' THEN 'February'
            WHEN '03' THEN 'March'
            WHEN '04' THEN 'April'
            WHEN '05' THEN 'May'
            WHEN '06' THEN 'June'
            WHEN '07' THEN 'July'
            WHEN '08' THEN 'August'
            WHEN '09' THEN 'September'
            WHEN '10' THEN 'October'
            WHEN '11' THEN 'November'
            WHEN '12' THEN 'December'
        END AS Month_Name,
        Landing_Outcome,
        Booster_Version,
        Launch_Site
    FROM SPACEXTABLE
    WHERE substr(Date, 0, 5) = '2017'
      AND Landing_Outcome = 'Success (ground pad)';
    """,
    con
)

df_check7
```
